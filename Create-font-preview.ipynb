```python
import numpy as np
from PIL import Image
import math
```

```python
characters = 190
characters_per_line = 26
image_width = characters_per_line * 12 #26 characters (items per line), 12 pixels per character
image_height = math.ceil(characters/characters_per_line) * 18 #18 pixels per character
image_height, image_width
```

```python
pixels = np.zeros(shape=(image_height, image_width, 3), dtype=np.uint8)
height, width = pixels.shape[:2]

file = open('fonts/default_font.c', 'r', encoding="utf-8")
line = file.readline()
pos_x = 0
pos_y = 0
while(len(line)!=0):
    if(not "// Code for char" in line):
        line = file.readline()
        continue
    last_comma_index = line.rfind(",  //")
    if last_comma_index == -1:
        last_comma_index = line.rfind(",") + 6
        

    line = line[:last_comma_index]
    line = line.strip()
    liste = list(map(lambda x: int(x, 16), line.split(",")))
    hex_array = np.array(liste[1:], dtype=np.uint8).reshape((10,2)).T
    binary_representation = np.unpackbits(hex_array, axis=0, bitorder='little')

    (y_indices, x_indices) = np.where(binary_representation == 1)

    pixels[  pos_y+y_indices, pos_x+x_indices] = (255, 255, 0)
    
    pos_x += 12
    if pos_x >= width-9:
        pos_x = 0
        pos_y += 18
    line = file.readline()
file.close()
new_img = Image.fromarray(pixels)
new_img.save('fonts/default_font_preview.png')
```
